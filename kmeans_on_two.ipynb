```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

def evaluate_kmeans(X, k_values):
    distances = []

    for k in k_values:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        
        # Find the index of the nearest cluster center for each sample
        labels = kmeans.predict(X)
        
        # Find the distance of each sample to its nearest cluster center
        distances_to_center = pairwise_distances_argmin_min(X, kmeans.cluster_centers_)[1]
        
        # Sum of Euclidean distances for each example from its class center
        sum_distance = np.sum(distances_to_center)
        distances.append(sum_distance)

    return distances

def plot_evaluation(k_values, distances, dataset_name):
    plt.plot(k_values, distances, marker='o')
    plt.title(f'K-means Performance on {dataset_name} Dataset')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Sum of Euclidean Distances')
    plt.show()

# Load the Iris dataset
```

```python
iris_data = load_iris()
X_iris = iris_data.data
k_values_iris = range(1, 11)  # Test k from 1 to 10
distances_iris = evaluate_kmeans(X_iris, k_values_iris)
plot_evaluation(k_values_iris, distances_iris, 'Iris')
```

```python
# Load the Breast Cancer dataset
cancer_data = load_breast_cancer()
X_cancer = cancer_data.data
k_values_cancer = range(1, 11)  # Test k from 1 to 10
distances_cancer = evaluate_kmeans(X_cancer, k_values_cancer)
plot_evaluation(k_values_cancer, distances_cancer, 'BreastCancer')
```
